```python
from keras.models import load_model
import numpy as np
from tensorflow.keras.preprocessing import image
from IPython.display import Image
```

```python
model = load_model(r'C:\Users\Lenovo\Desktop\version final\cnn\Cnn.keras')
```

```python
def prepare_image(file):
    img_path = ''
    img = image.load_img(img_path + file, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = img_array / 255.0
    img_array_expanded_dims = np.expand_dims(img_array, axis=0)
    return img_array_expanded_dims
```

```python
Image(filename=r"C:\Users\Lenovo\Desktop\version final\evaluation\111.jpg", width=300,height=200) 
```

```python
preprocessed_image = prepare_image(r"C:\Users\Lenovo\Desktop\version final\evaluation\111.jpg")
predictions = model.predict(preprocessed_image)
prediction_value = predictions[0][0]

if prediction_value > 0.5:
    fire_percentage = prediction_value * 100
    nonfire_percentage = (1 - prediction_value) * 100
else:
    fire_percentage = prediction_value * 100
    nonfire_percentage = (1 - prediction_value) * 100

print("Fire :", round(fire_percentage, 2), "%" , "|" , "Non Fire :", round(nonfire_percentage, 2), "%")

if prediction_value > 0.5:
    print("Fire detected!")
else:
    print("No fire detected.")
```
